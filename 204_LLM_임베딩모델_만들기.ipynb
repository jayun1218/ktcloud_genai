### **ì˜ˆì œ: "ìš°ë¦¬ ì¤‘ì— ìŠ¤íŒŒì´ê°€ ìˆì–´!"** - ì´ìƒí•œ ë‹¨ì–´ ì°¾ê¸° ê²Œì„
ë‹¨ì–´ ë¦¬ìŠ¤íŠ¸ì—ì„œ í™€ë¡œ ë™ë–¨ì–´ì§„ ì˜ë¯¸ì˜ ë‹¨ì–´ë¥¼ ë²¡í„° ê³µê°„ì—ì„œì˜ ê±°ë¦¬ë¥¼ ì´ìš©í•´ ì°¾ëŠ” ê²Œì„

```python
import numpy as np
from sentence_transformers import SentenceTransformer

# ëª¨ë¸ ë¡œë“œ
model = SentenceTransformer('intfloat/multilingual-e5-base')

# ë‹¨ì–´ ë¦¬ìŠ¤íŠ¸ (ìŠ¤íŒŒì´ê°€ í•˜ë‚˜ ìˆ¨ì–´ìˆìŒ)
words = ["ì»´í“¨í„°", "ë…¸íŠ¸ë¶", "í‚¤ë³´ë“œ", "ëª¨ë‹ˆí„°", "ë°”ë‚˜ë‚˜"]

# ê° ë‹¨ì–´ ì„ë² ë”©
embeddings = model.encode(words)


# ê° ë‹¨ì–´ ë²¡í„° ê°„ì˜ í‰ê·  ê±°ë¦¬ ê³„ì‚°
# (ìì‹ ì„ ì œì™¸í•œ ë‹¤ë¥¸ ëª¨ë“  ë‹¨ì–´ì™€ì˜ í‰ê·  ê±°ë¦¬ê°€ ê°€ì¥ ë¨¼ ë‹¨ì–´ê°€ ìŠ¤íŒŒì´!)
distances = []
for i in range(len(embeddings)):
    # ië²ˆì§¸ ë‹¨ì–´ì™€ ë‹¤ë¥¸ ëª¨ë“  ë‹¨ì–´ ê°„ì˜ ìœ í´ë¦¬ë“œ ê±°ë¦¬ í•©ê³„
    dist_sum = sum([np.linalg.norm(embeddings[i] - embeddings[j]) for j in range(len(embeddings)) if i != j])
    distances.append(dist_sum)


# í‰ê·  ê±°ë¦¬ê°€ ê°€ì¥ ë¨¼ ë‹¨ì–´(ìŠ¤íŒŒì´) ì°¾ê¸°
spy_index = np.argmax(distances)

print(f"ë‹¨ì–´ ë¦¬ìŠ¤íŠ¸: {words}")
print(f"ìŠ¤íŒŒì´ëŠ” ë°”ë¡œ... '{words[spy_index]}'(ì´)ì•¼!")

# ê²°ê³¼ ì˜ˆì¸¡:
# ìŠ¤íŒŒì´ëŠ” ë°”ë¡œ... 'ë°”ë‚˜ë‚˜'(ì´)ì•¼!
```

### **ì†ì‹¤ í•¨ìˆ˜ ì •ì˜**
- ì†ì‹¤ í•¨ìˆ˜(**Loss Function**):

- **ë‹¤ì–‘í•œ ì†ì‹¤í•¨ìˆ˜**(for sentence-transformers) :
    - https://www.sbert.net/docs/package_reference/sentence_transformer/losses.html
- **SoftmaxLoss** :  
    - https://www.sbert.net/docs/package_reference/sentence_transformer/losses.html#softmaxloss
    - **í•µì‹¬ ì•„ì´ë””ì–´**:
        - **"ë¶„ë¥˜(Classification) ë¬¸ì œë¡œ ë°”ê¿”ì„œ í’€ê¸°"**
    - ì´ ë°©ì‹ì€ ë‘ ë¬¸ì¥ì„ ì…ë ¥ë°›ì•„ ë‘˜ì˜ ê´€ê³„ê°€ 'ìœ ì‚¬í•œ ê´€ê³„(entailment)'ì¸ì§€, 'ìƒë°˜ëœ ê´€ê³„(contradiction)'ì¸ì§€ ë¶„ë¥˜í•˜ëŠ” ë¬¸ì œë¥¼ í‘¼ë‹¤.ëª¨ë¸ì´ ì´ ë¶„ë¥˜ ë¬¸ì œë¥¼ ì˜ í’€ê²Œ ë˜ë©´, ê·¸ ê³¼ì •ì—ì„œ ë¬¸ì¥ì˜ ì˜ë¯¸ë¥¼ ì˜ ì´í•´í•˜ê²Œ ë˜ê³  ìì—°ìŠ¤ëŸ½ê²Œ ì¢‹ì€ ì„ë² ë”©ì´ ë§Œë“¤ì–´ì§„ë‹¤ê³  ë´„
    
    - **í•™ìŠµ ê³¼ì •** :
        1. (A, P) ìŒì„ ëª¨ë¸ì— ë„£ì–´ì£¼ê³  "ì´ ë‘˜ì€ ìœ ì‚¬í•œ ê´€ê³„ì•¼ (ì •ë‹µ: 1)" ë¼ê³  ì•Œë ¤ì¤ë‹ˆë‹¤.
        2. (A, N) ìŒì„ ëª¨ë¸ì— ë„£ì–´ì£¼ê³  "ì´ ë‘˜ì€ ìƒë°˜ëœ ê´€ê³„ì•¼ (ì •ë‹µ: 0)" ë¼ê³  ì•Œë ¤ì¤ë‹ˆë‹¤.
        3. ëª¨ë¸ì€ ì´ 'ì •ë‹µ ë§íˆê¸°' í€´ì¦ˆë¥¼ ê³„ì† í’€ë©´ì„œ ë¬¸ì¥ì˜ ë¯¸ë¬˜í•œ ì˜ë¯¸ ì°¨ì´ë¥¼ í•™ìŠµí•©ë‹ˆë‹¤.
        - ìš”ì•½: ë‘ ë¬¸ì¥ì˜ ê´€ê³„ë¥¼ ë§íˆëŠ” 'ë¶„ë¥˜' ì‘ì—…ì„ í†µí•´ ë¬¸ì¥ ì„ë² ë”© ëŠ¥ë ¥ì„ ê°„ì ‘ì ìœ¼ë¡œ í•™ìŠµì‹œí‚µë‹ˆë‹¤.


```python
# 3) ì†ì‹¤ í•¨ìˆ˜
from sentence_transformers import losses

# ì†Œí”„íŠ¸ë§¥ìŠ¤ ì†ì‹¤ì„ ìœ„í•´ ëª…ì‹œì ìœ¼ë¡œ ë ˆì´ë¸”ì˜ ê°œìˆ˜ë¥¼ ì§€ì •í•´ì•¼ í•©ë‹ˆë‹¤.
train_loss = losses.SoftmaxLoss(
    model=embedding_model,
    sentence_embedding_dimension=embedding_model.get_sentence_embedding_dimension(),
    num_labels=3    # ë ˆì´ë¸” ê°œìˆ˜
)
```

### **í‰ê°€ì ì •ì˜**

- **Evaluator ê°ì²´**:
    - í›ˆë ¨ ê³¼ì •ì—ì„œ ëª¨ë¸ì„ í‰ê°€í•˜ê³  ë””ìŠ¤í¬ì— ì €ì¥í•  ìµœìƒì˜ ëª¨ë¸ì„ ê²°ì •í•˜ê¸° ìœ„í•´ Evaluator ê°ì²´ ì •ì˜
- **STS-B**(Semantic Textual Similarity Benchmark STS-B):
    - ëª¨ë¸ ì„±ëŠ¥ í‰ê°€ ë²¤ì¹˜ë§ˆí¬, **ë¬¸ì¥ ìœ ì‚¬ë„ ì¸¡ì •ê¸°**

```python
# 4) í‰ê°€ì: STS-B (ì½”ì‚¬ì¸ ë“± ì—¬ëŸ¬ ìœ ì‚¬ë„)
from sentence_transformers.evaluation import EmbeddingSimilarityEvaluator

val_sts = load_dataset('glue', 'stsb', split='validation')

evaluator = EmbeddingSimilarityEvaluator(
    sentences1=val_sts["sentence1"],
    sentences2=val_sts["sentence2"],
    scores=[score/5 for score in val_sts["label"]],   # 0~1 ì‚¬ì´ ê°’ìœ¼ë¡œ ì¡°ì •
    main_similarity="cosine",
    similarity_fn_names=["cosine", "euclidean", "manhattan", "dot"]
)
```

```python
# (êµì¬ ì¶”ì²œ ì½”ë“œì´ë‚˜ ì˜¤ë¥˜ ë°œìƒìœ¼ë¡œ ì•„ë˜ ì½”ë“œë¡œ ëŒ€ì²´í•¨)
# ì˜¤ë¥˜ ì›ì¸: MTEB, pydantic íƒœìŠ¤í¬ ë©”íƒ€ë°ì´í„° í˜¸í™˜ì„± ì˜¤ë¥˜
# from mteb import MTEB

# # í‰ê°€ ì‘ì—…ì„ ì„ íƒí•©ë‹ˆë‹¤.
# evaluation = MTEB(tasks=["Banking77Classification"])

# # ê²°ê³¼ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤.
# results = evaluation.run(embedding_model)
# results
```

```python
# MTEB, pydantic íƒœìŠ¤í¬ ë©”íƒ€ë°ì´í„° í˜¸í™˜ì„± ì¼ì¹˜ì‹œí‚´
import mteb
import pydantic
from sentence_transformers import SentenceTransformer

embedding_model = SentenceTransformer("bert-base-uncased")

# 1. í‰ê°€í•˜ë ¤ëŠ” ëª¨ë“  íƒœìŠ¤í¬ì˜ 'ì´ë¦„'ì„ ë¦¬ìŠ¤íŠ¸ë¡œ ì •ì˜í•©ë‹ˆë‹¤.
task_names = ["Banking77Classification"]


# 2. íŒ¨ì¹˜ê°€ ì™„ë£Œëœ íƒœìŠ¤í¬ 'ê°ì²´'ë“¤ì„ ë‹´ì„ ë¹ˆ ë¦¬ìŠ¤íŠ¸ë¥¼ ìƒì„±í•©ë‹ˆë‹¤.
patched_tasks_to_run = []


print("MTEB í‰ê°€ ì „, íƒœìŠ¤í¬ ë©”íƒ€ë°ì´í„° í˜¸í™˜ì„± íŒ¨ì¹˜ë¥¼ ì‹œì‘í•©ë‹ˆë‹¤...")
# 3. ë¦¬ìŠ¤íŠ¸ì— ìˆëŠ” ëª¨ë“  íƒœìŠ¤í¬ ì´ë¦„ì— ëŒ€í•´ ë°˜ë³µ ì‘ì—…ì„ ìˆ˜í–‰í•©ë‹ˆë‹¤.
for name in task_names:
    # 3-1. ì´ë¦„ìœ¼ë¡œ íƒœìŠ¤í¬ 'ê°ì²´'ë¥¼ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤.
    task = mteb.get_task(name)

    # 3-2. pydantic ë²„ì „ì— ë§ì¶° ë©”íƒ€ë°ì´í„°ë¥¼ ë”•ì…”ë„ˆë¦¬ë¡œ ë³€í™˜í•©ë‹ˆë‹¤.
    if pydantic.VERSION.startswith("1."):
        meta_dict = task.metadata.to_dict()
    else:
        meta_dict = task.metadata.model_dump()

    # 3-3. 'languages' í‚¤ê°€ ìˆëŠ”ì§€ ë°©ì–´ì ìœ¼ë¡œ í™•ì¸í•˜ê³  ì²˜ë¦¬í•©ë‹ˆë‹¤.
    if 'languages' not in meta_dict or not meta_dict['languages']:
        # í‚¤ê°€ ì—†ê±°ë‚˜ ë¦¬ìŠ¤íŠ¸ê°€ ë¹„ì–´ìˆìœ¼ë©´, ìƒˆë¡œ ì¶”ê°€/ì„¤ì •í•©ë‹ˆë‹¤.
        print(f"  - '{name}' íƒœìŠ¤í¬ì— 'languages' í‚¤ê°€ ì—†ì–´ ['en-en']ìœ¼ë¡œ ì„¤ì •í•©ë‹ˆë‹¤.")
        meta_dict['languages'] = ['en-en']
    else:
        # í‚¤ê°€ ìˆìœ¼ë©´, í˜•ì‹ë§Œ ìˆ˜ì •í•©ë‹ˆë‹¤.
        original_langs = meta_dict['languages']
        patched_langs = []
        for lang in original_langs:
            if '-' not in lang:
                patched_langs.append(f"{lang}-{lang}")
            else:
                patched_langs.append(lang)

        if original_langs != patched_langs:
            print(f"  - '{name}' íƒœìŠ¤í¬ íŒ¨ì¹˜: {original_langs} -> {patched_langs}")
            meta_dict['languages'] = patched_langs

    # 3-4. ìˆ˜ì •ëœ ë”•ì…”ë„ˆë¦¬ë¡œ ë©”íƒ€ë°ì´í„° ê°ì²´ë¥¼ ìƒˆë¡œ ë§Œë“¤ì–´ êµì²´í•©ë‹ˆë‹¤.
    task.metadata = task.metadata.__class__(**meta_dict)

    # 3-5. íŒ¨ì¹˜ê°€ ì™„ë£Œëœ íƒœìŠ¤í¬ ê°ì²´ë¥¼ ìµœì¢… ë¦¬ìŠ¤íŠ¸ì— 'ë°˜ë“œì‹œ' ì¶”ê°€í•©ë‹ˆë‹¤.
    patched_tasks_to_run.append(task)

# 4. 'íŒ¨ì¹˜ ì™„ë£Œëœ ê°ì²´ ë¦¬ìŠ¤íŠ¸'ë¥¼ MTEBì— ì „ë‹¬í•˜ì—¬ í‰ê°€ ê°ì²´ë¥¼ ìƒì„±í•©ë‹ˆë‹¤.
#    ì´ì œ ë¦¬ìŠ¤íŠ¸ê°€ ë¹„ì–´ìˆì§€ ì•Šìœ¼ë¯€ë¡œ IndexErrorê°€ ë°œìƒí•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.
print("\nëª¨ë“  íƒœìŠ¤í¬ íŒ¨ì¹˜ ì™„ë£Œ. MTEB í‰ê°€ë¥¼ ì‹œì‘í•©ë‹ˆë‹¤.")
evaluation = mteb.MTEB(tasks=patched_tasks_to_run)

# 5. í‰ê°€ë¥¼ ì‹¤í–‰í•©ë‹ˆë‹¤.
results = evaluation.run(
    embedding_model,
    output_folder="results/bert-base-uncased-final",
    eval_splits=["test"]
)

# 6. ê²°ê³¼ í™•ì¸
print(results)
```

```python
# 1) ë°ì´í„° ë¡œë“œ
from datasets import Dataset, load_dataset

# GLUEë¡œë¶€í„° MNLI ë°ì´í„°ì…‹ì„ ë¡œë“œí•©ë‹ˆë‹¤.
# 0 = ìˆ˜ë°˜, 1 = ì¤‘ë¦½, 2 = ëª¨ìˆœ
train_dataset = load_dataset("glue", "mnli", split="train").select(range(50_000))
train_dataset = train_dataset.remove_columns("idx")

# (ì¤‘ë¦½/ëª¨ìˆœ)=0, (ìˆ˜ë°˜)=1
mapping = {0: 1, 1: 0, 2: 0}
train_dataset = Dataset.from_dict({
    "sentence1": train_dataset["premise"],      # ì „ì œ
    "sentence2": train_dataset["hypothesis"],   # ê°€ì„¤
    "label": [float(mapping[label]) for label in train_dataset["label"]]
})
train_dataset[0]
```

```python
# 2) í‰ê°€ì ì •ì˜
from sentence_transformers.evaluation import EmbeddingSimilarityEvaluator

# STSBë¥¼ ìœ„í•œ ì„ë² ë”© ìœ ì‚¬ë„ í‰ê°€ìë¥¼ ë§Œë“­ë‹ˆë‹¤.
val_sts = load_dataset('glue', 'stsb', split='validation')
evaluator = EmbeddingSimilarityEvaluator(
    sentences1=val_sts["sentence1"],
    sentences2=val_sts["sentence2"],
    scores=[score/5 for score in val_sts["label"]],
    main_similarity="cosine",
    similarity_fn_names=["cosine", "euclidean", "manhattan", "dot"]
)
```

```python
# 3) ëª¨ë¸ í›ˆë ¨ : ëª¨ë¸ ì •ì˜ + ì†ì‹¤ í•¨ìˆ˜ + í›ˆë ¨ ë§¤ê°œë³€ìˆ˜ + ëª¨ë¸ í›ˆë ¨
from sentence_transformers import losses, SentenceTransformer
from sentence_transformers.trainer import SentenceTransformerTrainer
from sentence_transformers.training_args import SentenceTransformerTrainingArguments

# ëª¨ë¸
embedding_model = SentenceTransformer('bert-base-uncased')

# ì†ì‹¤ í•¨ìˆ˜
train_loss = losses.CosineSimilarityLoss(model=embedding_model)

# í›ˆë ¨ ë§¤ê°œë³€ìˆ˜
args = SentenceTransformerTrainingArguments(
    output_dir="cosineloss_embedding_model",
    num_train_epochs=1,
    per_device_train_batch_size=32,
    per_device_eval_batch_size=32,
    warmup_steps=100,
    fp16=True,
    eval_steps=100,
    logging_steps=100,
    report_to=[]
)

# ëª¨ë¸ í›ˆë ¨
trainer = SentenceTransformerTrainer(
    model=embedding_model,
    args=args,
    train_dataset=train_dataset,
    loss=train_loss,
    evaluator=evaluator
)
trainer.train()
```

#### **MNR ì†ì‹¤** (**Multiple Negatives Ranking Loss**)

    - ì§ˆë¬¸ê³¼ ê´€ë ¨ì—†ëŠ” ë‹µë³€ì€ ê±°ë¦¬ê°€ ìµœëŒ€í™”(ì˜ë¯¸ê°€ ë‹¤ë¥¸ ì—¬ëŸ¬ ë¶€ì • ìŒê³¼ëŠ” **ë°€ì–´ë‚´ë„ë¡**(push) )í•˜ì—¬ ì„ë² ë”© ê³µê°„ì„ íš¨ê³¼ì ìœ¼ë¡œ í•™ìŠµ

    1. ë¯¸ë‹ˆ ë°°ì¹˜ì— (A1, P1), (A2, P2), ..., (An, Pn)ê³¼ ê°™ì´ ì—¬ëŸ¬ ìŒì„ ë„£ìŠµë‹ˆë‹¤.
    2. A1ì„ ê¸°ì¤€ìœ¼ë¡œ, P1ê³¼ì˜ ìœ ì‚¬ë„ ì ìˆ˜ê°€ ë‚˜ë¨¸ì§€ ëª¨ë“  Positive(P2ë¶€í„° Pnê¹Œì§€)ì™€ì˜ ìœ ì‚¬ë„ ì ìˆ˜ë³´ë‹¤ ì›”ë“±íˆ ë†’ê²Œ ë‚˜ì˜¤ë„ë¡ ëª¨ë¸ì„ í•™ìŠµì‹œí‚µë‹ˆë‹¤. ì¦‰, A1ì—ê²Œ P1ì´ ì •ë‹µì´ê³  ë‚˜ë¨¸ì§€ëŠ” ëª¨ë‘ ì˜¤ë‹µ(Negative)ì´ë¼ê³  ì•Œë ¤ì£¼ëŠ” ê²ƒê³¼ ê°™ìŠµë‹ˆë‹¤.
    - ìš”ì•½: "ë³´ê¸° ì—¬ëŸ¬ ê°œ ì¤‘ì— ì •ë‹µ ê³ ë¥´ê¸°" ì‹œí—˜ì„ ì¹˜ë¥´ê²Œ í•˜ì—¬, ëª¨ë¸ì´ í›¨ì”¬ ë” ì •êµí•˜ê²Œ ë¬¸ì¥ ìŒì„ êµ¬ë¶„í•˜ë„ë¡ ë§Œë“­ë‹ˆë‹¤.

```python
# 1) ë°ì´í„° ë¡œë“œ
import random
from tqdm import tqdm
from datasets import Dataset, load_dataset

# GLUEë¡œë¶€í„° MNLI ë°ì´í„°ì…‹ì„ ë¡œë“œí•©ë‹ˆë‹¤.
# 0 = ìˆ˜ë°˜, 1 = ì¤‘ë¦½, 2 = ëª¨ìˆœ
mnli = load_dataset("glue", "mnli", split="train").select(range(50_000))
mnli = mnli.remove_columns("idx")

# MNLIì˜ ë ˆì´ë¸” ì¤‘ 0(ìˆ˜ë°˜ ê´€ê³„:ì „ì œì™€ ê°€ì„¤ì´ ì„œë¡œ ì˜ë¯¸ìƒ ì—°ê´€ëœ ìƒ˜í”Œ)ë§Œ
# 'ê¸ì • ìŒ(positive pair)'ìœ¼ë¡œ ì§€ì •
mnli = mnli.filter(lambda x: True if x['label'] == 0 else False)

# ë°ì´í„°ë¥¼ ì¤€ë¹„í•©ë‹ˆë‹¤.
# ì‚¼ì¤‘í•­(anchor:ë‹¤ë¥¸ ë¬¸ì¥ê³¼ ë¹„êµí•˜ê¸° ìœ„í•´ ì‚¬ìš©)
train_dataset = {"anchor": [], "positive": [], "negative": []}

soft_negatives = mnli["hypothesis"]
random.shuffle(soft_negatives)
for row, soft_negative in tqdm(zip(mnli, soft_negatives)):
    train_dataset["anchor"].append(row["premise"])
    train_dataset["positive"].append(row["hypothesis"])
    train_dataset["negative"].append(soft_negative)

train_dataset = Dataset.from_dict(train_dataset)
len(train_dataset)
```

```python
# zip ê°ì²´ì˜ ê¸¸ì´ë¥¼ í™•ì¸í•˜ëŠ” ë°©ë²• (zip ê°ì²´ë¥¼ ë¦¬ìŠ¤íŠ¸ë¡œ ë³€í™˜í•˜ì—¬ ê¸¸ì´ë¥¼ ê³„ì‚°)
print(f"zip ê°ì²´ì˜ ê°œìˆ˜: {len(list(zip(mnli, soft_negatives)))}")
```

```python
# 2) í‰ê°€ì ì •ì˜
from sentence_transformers.evaluation import EmbeddingSimilarityEvaluator

# STS-Bë¥¼ ìœ„í•´ ì„ë² ë”© ìœ ì‚¬ë„ í‰ê°€ìë¥¼ ë§Œë“­ë‹ˆë‹¤.
val_sts = load_dataset('glue', 'stsb', split='validation')
evaluator = EmbeddingSimilarityEvaluator(
    sentences1=val_sts["sentence1"],
    sentences2=val_sts["sentence2"],
    scores=[score/5 for score in val_sts["label"]],
    main_similarity="cosine",
    similarity_fn_names=["cosine", "euclidean", "manhattan", "dot"]
)
```

```python
# 3) ëª¨ë¸ í›ˆë ¨ : ëª¨ë¸ ì •ì˜ + ì†ì‹¤ í•¨ìˆ˜ + í›ˆë ¨ ë§¤ê°œë³€ìˆ˜ + ëª¨ë¸ í›ˆë ¨
from sentence_transformers import losses, SentenceTransformer
from sentence_transformers.trainer import SentenceTransformerTrainer
from sentence_transformers.training_args import SentenceTransformerTrainingArguments

# ëª¨ë¸
embedding_model = SentenceTransformer('bert-base-uncased')

# ì†ì‹¤ í•¨ìˆ˜
train_loss = losses.MultipleNegativesRankingLoss(model=embedding_model)

# í›ˆë ¨ ë§¤ê°œë³€ìˆ˜
args = SentenceTransformerTrainingArguments(
    output_dir="mnrloss_embedding_model",
    num_train_epochs=1,
    per_device_train_batch_size=32,
    per_device_eval_batch_size=32,
    warmup_steps=100,
    fp16=True,
    eval_steps=100,
    logging_steps=100,
    report_to=[]
)

# ëª¨ë¸ í›ˆë ¨
trainer = SentenceTransformerTrainer(
    model=embedding_model,
    args=args,
    train_dataset=train_dataset,
    loss=train_loss,
    evaluator=evaluator
)
trainer.train()
```

```python
# í›ˆë ¨ëœ ëª¨ë¸ì„ í‰ê°€í•©ë‹ˆë‹¤.
evaluator(embedding_model)

# MultipleNegativesRankingLoss ì†ì‹¤í•¨ìˆ˜ ì‚¬ìš© ê²°ê³¼ :
# - ì•ì—ì„œ ì‚¬ìš©í•œ SoftmaxLoss ê²°ê³¼('pearson_cosine': 0.533716951543377)
# - ì•ì—ì„œ ì‚¬ìš©í•œ CosineSimilarityLoss ê²°ê³¼ ('pearson_cosine': 0.7142764856270979)
# - ë³´ë‹¤ ëª¨ë¸ì˜ ìœ ì‚¬ë„ê°€ í–¥ìƒë¨ ('pearson_cosine': 0.8017878954247927)
```
