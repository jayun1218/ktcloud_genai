```python
from transformers import AutoTokenizer
from datasets import load_dataset

# ì±„íŒ… í…œí”Œë¦¿ì„ ì‚¬ìš©í•˜ê¸° ìœ„í•´ í† í¬ë‚˜ì´ì €ë¥¼ ë¡œë“œí•©ë‹ˆë‹¤.
template_tokenizer = AutoTokenizer.from_pretrained("TinyLlama/TinyLlama-1.1B-Chat-v1.0")


def format_prompt(example):
    """TinyLlamaì˜ <|user|> í…œí”Œë¦¿ìœ¼ë¡œ í”„ë¡¬í”„íŠ¸ë¥¼ í¬ë§·íŒ…í•©ë‹ˆë‹¤"""

    # ì±„íŒ… í…œí”Œë¦¿ êµ¬ì„±
    chat = example["messages"]
    prompt = template_tokenizer.apply_chat_template(chat, tokenize=False)

    return {"text": prompt}


# ë°ì´í„°ë¥¼ ë¡œë“œí•˜ê³  TinyLlama í…œí”Œë¦¿ì„ ì ìš©í•©ë‹ˆë‹¤.
dataset = (
    load_dataset("HuggingFaceH4/ultrachat_200k",  split="test_sft")
      .shuffle(seed=42)
      .select(range(3_000))
)
dataset = dataset.map(format_prompt).remove_columns(['messages'])
```

### **ëª¨ë¸ ì–‘ìí™”**

- **bitsandbytes** íŒ¨í‚¤ì§€ë¥¼ ì‚¬ìš©í•´ ì‚¬ì „ í›ˆë ¨ ëª¨ë¸ì„ 4ë¹„íŠ¸ í‘œí˜„ìœ¼ë¡œ ì••ì¶•í•¨
    - í•˜ë“œì›¨ì–´ ì‚¬ì–‘ ì°¸ê³ : https://github.com/bitsandbytes-foundation/bitsandbytes
    
- **BitsAndBytesConfig** :  4-ë¹„íŠ¸ ì–‘ìí™” ì„¤ì •
    - ì˜ˆ: 4GB VRAM --> 1GB VRAM

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig

model_name = "TinyLlama/TinyLlama-1.1B-intermediate-step-1431k-3T"

# 4-ë¹„íŠ¸ ì–‘ìí™” ì„¤ì • - QLoRAì˜ Q ë‹¨ê³„
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,  # 4-ë¹„íŠ¸ ì •ë°€ë„ ëª¨ë¸ ë¡œë“œ
    bnb_4bit_quant_type="nf4",  # ì–‘ìí™” ì¢…ë¥˜
    bnb_4bit_compute_dtype="float16",  # ê³„ì‚° dtype
    bnb_4bit_use_double_quant=True,  # ì´ì¤‘ ì–‘ìí™” ì ìš©
)

# ëª¨ë¸ì„ ë¡œë“œí•˜ê³  GPUì—ì„œ í›ˆë ¨í•©ë‹ˆë‹¤.
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto",

    # ì¼ë°˜ì ì¸ SFTì—ì„œëŠ” ë‹¤ìŒì„ ì‚­ì œí•˜ì„¸ìš”.
    quantization_config=bnb_config,
)
model.config.use_cache = False

# LLaMA í† í¬ë‚˜ì´ì € ë¡œë“œ
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
tokenizer.pad_token = "<PAD>"
tokenizer.padding_side = "left"
```

### **ì„¤ì •**

#### **LoRA ì„¤ì •**
- **peft** :
    - ê±°ëŒ€ AI ëª¨ë¸ì˜ ëª¨ë“  íŒŒë¼ë¯¸í„°ë¥¼ ê±´ë“œë¦¬ì§€ ì•Šê³  LoRAì²˜ëŸ¼ ì•„ì£¼ ì‘ì€ ì¼ë¶€ íŒŒë¼ë¯¸í„°ë§Œ ì¶”ê°€í•˜ê±°ë‚˜ ìˆ˜ì •í•˜ì—¬, ì ì€ ë©”ëª¨ë¦¬ì™€ ì»´í“¨íŒ… ìì›ìœ¼ë¡œë„ ë¹ ë¥´ê³  íš¨ìœ¨ì ì¸ ë¯¸ì„¸ì¡°ì •(fine-tuning)ì„ ê°€ëŠ¥í•˜ê²Œ í•´ì£¼ëŠ” í—ˆê¹…í˜ì´ìŠ¤ì˜ ë¼ì´ë¸ŒëŸ¬ë¦¬
    - LoRA  ì„¤ì • êµ¬ì„± --> ë¯¸ì„¸ íŠœë‹ ê³¼ì •ì˜ í•˜ì´í¼íŒŒë¼ë¯¸í„°ì— í•´ë‹¹
    - https://github.com/huggingface/peft
    

- [ì°¸ê³ ] ë¯¸ì„¸ íŠœë‹ì— ëŒ€í•œ ì¶”ê°€ì ì¸ ì •ë³´

 - target_modules=  # ëŒ€ìƒ ì¸µ (adaptrorë¥¼ ì§€ì •í•  ì¸µ)
    - ['k_proj', 'gate_proj', 'v_proj', 'up_proj', 'q_proj', 'o_proj', 'down_proj']
    - LoRA(ì €ë­í¬ ì–´ëŒ‘í„°)ë¥¼ ì‚½ì…í•  â€œê°€ì¤‘ì¹˜ í–‰ë ¬(Linear layer)â€ì˜ ì´ë¦„ ëª©ë¡.
    ì¦‰, ì•„ë˜ì— ì ì€ ì´ë¦„ì„ ê°€ì§„ ì„ í˜•ì¸µë“¤ì—ë§Œ LoRAì˜ ì €ë­í¬ ì—…ë°ì´íŠ¸ë¥¼ ë¶™ì—¬ì„œ, ê·¸ ë¶€ë¶„ë§Œ í•™ìŠµ ê°€ëŠ¥(ë‚˜ë¨¸ì§€ëŠ” ë™ê²°)í•˜ê²Œ ë§Œë“­ë‹ˆë‹¤. ë¦¬ìŠ¤íŠ¸ì— ë¬´ì—‡ì„ ë„£ëŠëƒê°€ í•™ìŠµ í’ˆì§ˆÂ·ì†ë„Â·VRAMì„ ì¢Œìš°í•©ë‹ˆë‹¤.
    - íŠ¸ëœìŠ¤í¬ë¨¸ì—ì„œ ì˜ë¯¸ ìˆëŠ” í‘œí˜„ ë³€í™”ê°€ ì£¼ë¡œ ì¼ì–´ë‚˜ëŠ” ê³³ì´
        - â‘  ì–´í…ì…˜ì˜ Q/K/V/O íˆ¬ì˜(query/key/value/output projections)ê³¼
        - â‘¡ FFN(MLP)ì˜ ì—…/ë‹¤ìš´/ê²Œì´íŠ¸ íˆ¬ì˜ ì´ë¯€ë¡œ, ì´ë“¤ì— LoRAë¥¼ ê±°ëŠ” ê²ƒì´ ê´€í–‰ì…ë‹ˆë‹¤.

```python
from peft import LoraConfig, prepare_model_for_kbit_training, get_peft_model

# LoRA ì„¤ì •
peft_config = LoraConfig(
    lora_alpha=128,     # LoRA ìŠ¤ì¼€ì¼ë§(ì›ë³¸ ê°€ì¤‘ì¹˜ì— ì¶”ê°€ë˜ëŠ” ë³€í™”ëŸ‰ì„ ì œì–´, ë³´í†µ r í¬ê¸°ì˜ 2ë°°)
    lora_dropout=0.1,   # LoRA ì¸µì˜ ë“œë¡­ì•„ì›ƒ
    r=64,               # ë­í¬(ì••ì¶•ëœ í–‰ë ¬ì˜ ë­í¬ 4~64)
    bias="none",
    task_type="CAUSAL_LM",
    target_modules=     # ëŒ€ìƒ ì¸µ (adaptrorë¥¼ ì§€ì •í•  ì¸µ)
     ['k_proj', 'gate_proj', 'v_proj', 'up_proj', 'q_proj', 'o_proj', 'down_proj']
)

# í›ˆë ¨ì„ ìœ„í•œ ëª¨ë¸ ì¤€ë¹„
model = prepare_model_for_kbit_training(model)
model = get_peft_model(model, peft_config)
```

#### **í›ˆë ¨ ì„¤ì •**
- í›ˆë ¨ ë§¤ê°œë³€ìˆ˜ ì„¤ì •
- **trl**
    - Transformer ëª¨ë¸ì„ ê°•í™”í•™ìŠµ(RLHF)ìœ¼ë¡œ fine-tuningí•˜ê¸° ìœ„í•œ Hugging Face ë¼ì´ë¸ŒëŸ¬ë¦¬

```python
from trl import SFTConfig

output_dir = "./results"

# í›ˆë ¨ ë§¤ê°œë³€ìˆ˜
training_arguments = SFTConfig(
    output_dir=output_dir,
    per_device_train_batch_size=2,
    gradient_accumulation_steps=4,
    optim="paged_adamw_32bit",      # QLoRA ë…¼ë¬¸ì—ì„œ ì‚¬ìš©ëœ í˜ì´ì§€ ê¸°ë°˜ ì˜µí‹°ë§ˆì´ì € ì§€ì •
    learning_rate=2e-4,             # ê°€ì¤‘ì¹˜ ì—…ë°ì´íŠ¸ ìŠ¤í… í¬ê¸°, í•™ìŠµë¥ , 0.0002
    lr_scheduler_type="cosine",     # í•™ìŠµë¥  ì¡°ì • ìŠ¤ì¼€ì¥´ëŸ¬, cosine(ë™ì ìœ¼ë¡œ í•™ìŠµë¥  ì¡°ì •í•¨)
    num_train_epochs=1,             # 1ë²ˆ í•™ìŠµ
    logging_steps=10,               # 10 ìŠ¤í…ë§ˆë‹¤ ì†ì‹¤ ì¶œë ¥
    fp16=True,
    gradient_checkpointing=True,
    dataset_text_field="text",
    max_length=512
)
```

### **í›ˆë ¨**

```python
# T4ì—ì„œ ì˜¤ë˜ ê±¸ë¦¼
from trl import SFTTrainer

# ì§€ë„ ë¯¸ì„¸ íŠœë‹ ë§¤ê°œë³€ìˆ˜ ì§€ì •
trainer = SFTTrainer(
    model=model,
    train_dataset=dataset,
    processing_class=tokenizer,
    args=training_arguments,

    # ì¼ë°˜ì ì¸ SFTì—ì„œëŠ” ë‹¤ìŒì„ ì‚­ì œí•˜ì„¸ìš”.
    peft_config=peft_config,
)

# ëª¨ë¸ í›ˆë ¨
trainer.train()

# QLoRA ê°€ì¤‘ì¹˜ ì €ì¥
trainer.model.save_pretrained("TinyLlama-1.1B-qlora")
```

```python
from datasets import load_dataset

def format_prompt(example):
    """TinyLlamaì˜ <|user|> í…œí”Œë¦¿ì„ ì‚¬ìš©í•´ í”„ë¡¬í”„íŠ¸ë¥¼ êµ¬ì„±í•©ë‹ˆë‹¤"""

    # í…œí”Œë¦¿ í¬ë§·íŒ…
    system = "<|system|>\n" + example['system'] + "</s>\n"
    prompt = "<|user|>\n" + example['input'] + "</s>\n<|assistant|>\n"
    chosen = example['chosen'] + "</s>\n"
    rejected = example['rejected'] + "</s>\n"

    return {
        "prompt": system + prompt,
        "chosen": chosen,
        "rejected": rejected,
    }

# ë°ì´í„°ì…‹ì— í…œí”Œë¦¿ì„ ì ìš©í•˜ê³  ë¹„êµì  ì§§ì€ ëŒ€ë‹µì„ ì„ íƒí•©ë‹ˆë‹¤
dpo_dataset = load_dataset("argilla/distilabel-intel-orca-dpo-pairs", split="train")
dpo_dataset = dpo_dataset.filter(
    lambda r:
        r["status"] != "tie" and
        r["chosen_score"] >= 8 and
        not r["in_gsm8k_train"]
)
dpo_dataset = dpo_dataset.map(format_prompt, remove_columns=dpo_dataset.column_names)
dpo_dataset
```

```python
from peft import AutoPeftModelForCausalLM
from transformers import BitsAndBytesConfig, AutoTokenizer

# 4-ë¹„íŠ¸ ì–‘ìí™” ì„¤ì • - QLoRAì˜ Q ë‹¨ê³„
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,          # 4-ë¹„íŠ¸ ì •ë°€ë„ ëª¨ë¸ ë¡œë“œ
    bnb_4bit_quant_type="nf4",  # ì–‘ìí™” ì¢…ë¥˜
    bnb_4bit_compute_dtype="float16",   # ê³„ì‚° dtype
    bnb_4bit_use_double_quant=True,     # ì´ì¤‘ ì–‘ìí™” ì ìš©
)

# LoRAì™€ ë² ì´ìŠ¤ ëª¨ë¸ì„ í•©ì¹©ë‹ˆë‹¤.
model = AutoPeftModelForCausalLM.from_pretrained(
    "./TinyLlama-1.1B-qlora",
    low_cpu_mem_usage=True,
    device_map="auto",
    quantization_config=bnb_config,
)
# merged_model = model.merge_and_unload()

# LLaMA í† í¬ë‚˜ì´ì €ë¥¼ ë¡œë“œí•©ë‹ˆë‹¤.
model_name = "TinyLlama/TinyLlama-1.1B-intermediate-step-1431k-3T"
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
tokenizer.pad_token = "<PAD>"
tokenizer.padding_side = "left"
```

```python
from transformers import pipeline

# ì¸ìŠ¤íŠ¸ëŸ­ì…˜ íŠœë‹ëœ ëª¨ë¸ì„ ì‹¤í–‰í•©ë‹ˆë‹¤.
pipe = pipeline(task="text-generation", model=dpo_model, tokenizer=tokenizer)

def create_dpo_prompt(user_input):
    """DPO í•™ìŠµ ëª¨ë¸ì„ ìœ„í•´ í”„ë¡¬í”„íŠ¸ë¥¼ í¬ë§·íŒ…í•©ë‹ˆë‹¤"""
    prompt = f"""<|user|>
{user_input}</s>
<|assistant|>
"""
    return prompt

def chat_with_dpo_model(user_input, max_new_tokens=256):
    """DPO ëª¨ë¸ì— ì…ë ¥ì„ ì£¼ê³  ì‘ë‹µì„ ë°›ìŠµë‹ˆë‹¤"""
    prompt = create_dpo_prompt(user_input)

    response = pipe(
        prompt,
        max_new_tokens=max_new_tokens,
        temperature=0.5,  # DPOëŠ” ë” ë‚®ì€ ì˜¨ë„ ê¶Œì¥
        top_p=0.9,
        do_sample=True,
        eos_token_id=tokenizer.eos_token_id,
    )

    # ìƒì„±ëœ ì „ì²´ í…ìŠ¤íŠ¸ì—ì„œ ëª¨ë¸ì˜ ì‘ë‹µ ë¶€ë¶„ë§Œ ì¶”ì¶œ
    generated_text = response[0]["generated_text"]

    # <|assistant|> ì´í›„ì˜ í…ìŠ¤íŠ¸ë§Œ ì¶”ì¶œ
    assistant_response = generated_text.split("<|assistant|>")[-1].strip()

    # </s> í† í° ì œê±°
    assistant_response = assistant_response.replace("</s>", "").strip()

    return assistant_response

# ì±—ë´‡ ë©”ì¸ ë£¨í”„
print("=" * 70)
print("ğŸ¤– DPO í•™ìŠµ TinyLlama í•œê¸€ ì±—ë´‡ì— ì˜¤ì‹  ê²ƒì„ í™˜ì˜í•©ë‹ˆë‹¤!")
print("=" * 70)
print("ğŸ’¡ íŒ: ëª…í™•í•œ ì§ˆë¬¸ì„ ì…ë ¥í•˜ë©´ ë” ì¢‹ì€ ë‹µë³€ì„ ë°›ì„ ìˆ˜ ìˆìŠµë‹ˆë‹¤.")
print("(ì¢…ë£Œí•˜ë ¤ë©´ 'exit' ë˜ëŠ” 'ì¢…ë£Œ'ë¥¼ ì…ë ¥í•˜ì„¸ìš”)\n")

conversation_count = 0

while True:
    # ì‚¬ìš©ì ì…ë ¥ ë°›ê¸°
    user_input = input("ğŸ‘¤ ë‹¹ì‹ : ").strip()

    # ì¢…ë£Œ ì¡°ê±´
    if user_input.lower() in ['exit', 'ì¢…ë£Œ', 'quit', 'q']:
        print("\nğŸ¤– ì±—ë´‡: ì•ˆë…•íˆ ê°€ì„¸ìš”! ë‹¤ìŒì— ë˜ ë§Œë‚˜ìš”. ğŸ‘‹")
        break

    # ë¹ˆ ì…ë ¥ ì²˜ë¦¬
    if not user_input:
        print("ğŸ¤– ì±—ë´‡: ì§ˆë¬¸ì„ ì…ë ¥í•´ ì£¼ì„¸ìš”.\n")
        continue

    # ëª¨ë¸ ì‘ë‹µ ìƒì„±
    print("\nğŸ¤– ì±—ë´‡: ", end="", flush=True)

    try:
        response = chat_with_dpo_model(user_input)
        print(response)
        print()

        conversation_count += 1
        if conversation_count % 5 == 0:
            print(f"ğŸ“Š {conversation_count}ë²ˆì˜ ëŒ€í™”ë¥¼ ë‚˜ëˆ´ìŠµë‹ˆë‹¤.\n")

    except Exception as e:
        print(f"ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤: {e}\n")
        continue
```
